```python
import numpy as np

N = 9 # the number of training data
X = np.array([[1.0,1.0,0.85,0.76],[1.0,1.0,0.0,0.24],[1.0,0.0,0.625,0.84],
              [1.0,0.0,0.45,0.96],[1.0,1.0,0.225,0.46],[1.0,0.0,0.25,0.7],
              [1.0,1.0,0.7,0.44],[1.0,0.0,0.3,0.4],[1.0,0.0,0.125,0.66]])
y = np.array([1,0,1,1,0,0,1,0,1])
w = np.array([0,0,0,0])
learning_rate = 0.1
iteration = 100

def sigmoid(z):   # sigmoid
    return 1 / (1 + np.exp(-z))

def loss_function(params, data_x, class_y):  # cross entropy loss
    fx = sigmoid(data_x.dot(params)) # y_hat 추정값 계산
    loss = np.sum(((-1) * class_y * np.log(fx)) - ((1 - class_y) * np.log(1 - fx)))  # loss 계산
    print(f'loss = {loss}')

# learning by gradient descent
for i in range(iteration):

    ### Fill In Your Code Here #################

    # compute y_hat from X and w
    y_hat = sigmoid(X.dot(w))
    # compute binary cross entropy loss and print
    loss_function(w, X, y)
    error = y - y_hat
    # update w using X, y, y_hat
    w = w + learning_rate * (X.T.dot(error))
    ############################################
    
# the learned model parameter
print(f'w = {w}\n')

# new data x_tom for tom = (sex=M, age=24, income=2500)
x_tom = np.array([1,1,0.1,0.5])
print(f'x_tom = {x_tom}')

### Fill In Your Code Here #################

# compute y_tom_hat and print
expected_tom = sigmoid(x_tom.dot(w))
print(f'y_tom_hat = {expected_tom}')
# print classification result
if expected_tom > 0.5:
    print('tom is (+)')
else:
    print('tom is (-)')
############################################
```
